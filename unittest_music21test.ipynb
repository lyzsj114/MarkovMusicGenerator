```python
def score2notelist(score):
    note_list = []
    for i in score.flatten():
        if isinstance(i, note.Note):
            name = i.name
            if i.name == 'C#':
                name = 'D-'
            elif i.name == 'D#':
                name = 'E-'
            elif i.name == 'G-':
                name = 'F#'
            elif i.name == 'G#':
                name = 'A-'
            elif i.name == 'A#':
                name = 'B-'
            elif i.name =='F##':
                name = 'G'
            elif i.name == 'B#':
                name = 'C'
            elif i.name == 'E#':
                name = 'F'
            present = (name+str(i.octave), i.duration.quarterLength)
            note_list.append(present)
    return note_list
```

```python
all_note_list = []
for i in range(len(bach)):
    note_list = score2notelist(bach[i].parse().parts[0])
    all_note_list.extend(note_list)
```

```python
note_set = set([i for i,_ in all_note_list])
```

```python
count = 0
new_note_set = []
for n in note_set:
    num = all_pits.count(n)
    if num > 20:
        count = count+1
        new_note_set.append(n)
        print(count, n, num)
```

```python
def rule(x):
    ET = {'C':1, 'D-':2, 'D':3, 'E-':4, 'E':5, 'F':6, 'F#':7, 'G':8, 'A-':9, 'A':10, 'B-':11, 'B':12}
    return int(x[-1])*13 + ET[x[:-1]]
```

```python
new_note_set = sorted(new_note_set, key=rule)
```

```python
f_note = {s:i for i,s in enumerate(new_note_set)}
```

```python
f_note.keys()
```

```python
prev = None
all_pairs = []
for i in range(len(bach)):
    score = bach[i].parse().parts[0]
    for n in score2notelist(score):
        present = n
        if prev is None:
            prev = present
            continue
        pair = (present, prev)
        prev = present
        all_pairs.append(pair)
```

```python
# Z
Z = np.zeros((25,25))
for p in all_pairs:
    first, second = p
    keys = f_note.keys()
    if first[0] not in keys or second[0] not in keys:
        continue
    i = f_note[first[0]]
    j = f_note[second[0]]
    Z[i,j] += 1
```

```python
sns.heatmap(Z, xticklabels=new_note_set, yticklabels=new_note_set)
```

```python
P = (Z.T / np.sum(Z,axis=1)).T
sns.heatmap(P, xticklabels=new_note_set, yticklabels=new_note_set)
```
